**#3** Web scrape the Champaign County Assessor's [website](https://champaignil.devnetwedge.com/) to extract a property's Net Taxable Value in US Dollars. Do this for the 1st, 2nd, 3rd, 4th, and 9th properties in the dataset from Problem **#2** and print these 5 values.


```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import numpy as np
root_url = 'https://champaignil.devnetwedge.com/parcel/view/'
net_tax02 = ['']*len(rentals)
for i in range(0,9):
    url = root_url + str(rentals['Parcel Number'].iloc[i]) + '/2020'
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    net_tax01 = soup.select('tr:nth-child(5) td:nth-child(1) .inner-value')
    for oN in net_tax01:
        net_tax02[i] = [taxes01.text for taxes01 in net_tax01]
net_tax03 = []
for item in net_tax02:
    net_tax03 += item
del net_tax03[4:8]
print(net_tax03)
```
